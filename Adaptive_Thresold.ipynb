Adaptive(mean and gaussian) thresolding.
This type of thresolding works making an image into several blocks and for everybloks they set autometically a threshold and works on it

```python

maximum_value = 255
ret, binary_thresh = cv.threshold(image, 80, maximum_value, cv.THRESH_BINARY)
adaptive_mean_threshold = cv.adaptiveThreshold(image, maximum_value, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)
adaptive_gaussian_threshold = cv.adaptiveThreshold(image, maximum_value, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)

titles = ['Original', 'Binary', 'Adap-Mean', 'Adap-Gaussian']
images = [image, binary_thresh, adaptive_mean_threshold, adaptive_gaussian_threshold]
for i in range(len(images)):
  plt.subplot(1, 4, i+1), plt.imshow(images[i], cmap='gray')
  plt.title(titles[i])
  plt.xticks([]), plt.yticks([])
plt.show()
```
