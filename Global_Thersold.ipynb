```python
threshold_value = 130
targeted_value = 255
ret, binary_thresh = cv.threshold(image, threshold_value, targeted_value, cv.THRESH_BINARY)
ret, binary_inv_thresh = cv.threshold(image, threshold_value, targeted_value, cv.THRESH_BINARY_INV)
ret, trunc_thresh = cv.threshold(image, threshold_value, targeted_value, cv.THRESH_TRUNC)
ret, tozero_thresh = cv.threshold(image, threshold_value, targeted_value, cv.THRESH_TOZERO)
ret, tozero_inv_thresh = cv.threshold(image, threshold_value, targeted_value, cv.THRESH_TOZERO_INV)

titles = ['main_image', 'binary_image', 'binary_inv_image', 'trunc_image', 'tozero_image', 'tozero_inv_image']
images = [image, binary_thresh, binary_inv_thresh, trunc_thresh, tozero_thresh, tozero_inv_thresh]
for i in range(len(images)):
  plt.subplot(2, 3, i+1), plt.imshow(images[i], cmap='gray')
  plt.title(titles[i])
  plt.xticks([]), plt.yticks([])
plt.show()
```
