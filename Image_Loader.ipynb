```python
import cv2 as cv
img = cv.imread("/content/drive/MyDrive/Computer_vision_dataset/Images/rose.jpg")
rgb_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(rgb_img)
```

```python
red, green, blue = cv.split(rgb_img)
titles = ['Original Image','Red Channel', 'Green Channel', 'Blue Channel']
images = [rgb_img, red, green, blue]
plt.figure(figsize=(10,16))
for i in range(4):
  plt.subplot(1,4,i+1)
  plt.imshow(images[i])
  plt.title(titles[i])
  plt.xticks([])
  plt.yticks([])
plt.show()
```
